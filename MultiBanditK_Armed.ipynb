```python
# Import necessary libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_runs = 1
num_steps = 2000
step = 0.2
q_values = [0,1,0,0]
arm_count = [1,0,0,0] 
epsilon = 0.1
last_action = 0
rewards = [0,1,4,2]
scores = [0]
averages = []
```

```python
def argmax(q_values):
    top = float("-inf")
    ties = [] 
    for i in range(len(q_values)):
        if q_values[i] > top:
            top, ties = q_values[i], [i]
        elif q_values[i] == top:
            ties.append(i)
    ind = np.random.choice(ties)    
    return ind

for i in range(num_steps):

  reward = rewards[last_action]
  arm_count[last_action] += 1
  q_values[last_action] += step * (reward - q_values[last_action])
  if np.random.random() < epsilon:
      current_action = np.random.randint(len(q_values))
  else:
      current_action = argmax(q_values)        
  last_action = current_action
  
  scores.append(scores[-1] + reward)
  averages.append(scores[-1] / (i + 1))
```

```python
plt.figure(figsize=(15, 5), dpi= 80, facecolor='w', edgecolor='k')

plt.plot([1.75 for _ in range(num_steps)], linestyle="--")
plt.plot(averages)
plt.legend(["A/B Testing Expected", "Multi-Bandit"])
plt.title("Average Reward of Epsilon Greedy Agent")
plt.xlabel("Steps")
plt.ylabel("Average reward")
plt.show()
```
